```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

```python
df=titanic.copy()
df
```

```python
pcl_grp=df.groupby('pclass')
for i, j in pcl_grp:
    print(i, '묶인 그룹')
    print(j, 'data')
# i는 그룹의 이름을 출력한다. 여기에서는 객실 이름
# j는 해당 객실 등급에 해당하는 승객들의 데이터만 모아놓은 작은 표를 출력한다.
# 그러므로 이렇게 할 시에 각 등급별 데이터만 출력하도록 된다
```

```python
#내가 원하는 그룹만 추출할 수 있다.
pcl_grp.get_group(3)
```

```python
# 2개 이상 컬럼을 묶기
pcl_grp2=df.groupby(['pclass','sex'])
for i,j in pcl_grp2:
    print(i,'묶인 그룹')
    print(j,'data')
```

```python
pcl_grp2.get_group((3,'female'))
```

```python
'''
데이터 병합
판다스에서 제공하는 merge, concat 함수

merge 기준 컬럼을 통한 ( 공통된 컬럼( 값 기준 )) 데이터를 병합한다.

공통된 기준 값인 컬럼이 필요하다
join 할 수 있는 방법이 다양하다 (left, right, outer, inner)
공통 컬럼 기준으로 여러 컬럼을 합칠 수 있다.
concat 데이터와 데이터 덩어리와 덩어리를 합쳐서 병합한다.

df1, df2를 예시로 붙인다면? 행과 열 기준으로 붙일 수 있다.
axis =0, 1에 따라서 붙이는 방법이 다르다.
df1 + df2 ....+ dfn (가로로 붙이는 경우)
(세로로 붙이는 경우 )
df1
'+'
df2
...
데이터 병합을 하는 이유는?

고객, 주문, 행사
나는 고객이 어떤 주문을하고 어떤 행사에 참여했는지를 보고 싶다.
그러면 고객+주문+행사를 합쳐서 봐야 한다.
'''
test1=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '학회원분반':['Python','SQL','R','데이터분석입문','데이터분석전처리']})
test2=pd.DataFrame({'학회원번호':[1,2,3,4,5],
             '상점현황':[4,3,5,10,2]})
display(test1)
display(test2)
```

```python
pd.merge(test1, test2, how='inner')
```

```python
# concat
pd.concat([test1,test2]) # 디폴트는 행기준으로 붙여진다.
```

```python
pd.concat([test1,test2],axis=1) # 디폴트는 행기준으로 붙여진다.
```
