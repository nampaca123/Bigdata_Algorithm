```python
import seaborn as sns
```

```python
tt = sns.load_dataset('titanic')
```

```python
# '생존'이며 '여성'이고 '1등석'을 이용하며 '쉐르부르에서 탑승'
tt.query('''survived == 1 and sex == "female" and pclass == 1 and embarked == "C"''')
```

```python
# '사망'이며 '남성'이고 '3등석'을 이용하며 '사우스햄튼에서 탑승'하고 '30세 이상'
tt.query('''survived == 0 and sex == "male" and pclass == 3 and embarked == "S" and age >= 30''')
```

```python
# '생존'이며 '남성'이고 '2등석'을 이용하며 '요금이 20파운드 이상'
tt.query('''survived == 1 and sex == "male" and pclass == 2 and fare >= 20''')
```

```python
# '여성'이며 '20세 이상 40세 미만'이고 '쉐르부르에서 탑승'하고 '2등석'을 이용
tt.query('''sex == "female" and age >= 20 and age < 40 and embarked == "C" and pclass == 2''')
```

```python
# '남성'이며 '1등석'을 이용하고 '요금이 50파운드 이상'이며 '40세 이상'
tt.query('''sex == "male" and pclass == 1 and fare >= 50 and age >= 40''')
```

```python
# '사망'이며 '여성'이고 '3등석'을 이용하며 '사우스햄튼에서 탑승'하고 '18세 미만'
tt.query('''survived == 0 and sex == "female" and pclass == 3 and embarked == "S" and age < 18''')
```

```python
# '생존'이며 '남성'이고 '50세 이상'이며 '1등석'을 이용하고 '요금이 30파운드 이상'
tt.query('''survived == 1 and sex == "male" and age >= 50 and pclass == 1 and fare >= 30''')
```

```python
# '여성'이며 '쉐르부르에서 탑승'하고 '1등석'을 이용하며 '요금이 100파운드 이상'
tt.query('''sex == "female" and embarked == "C" and pclass == 1 and fare >= 100''')
```

```python
# '남성'이며 '20세 미만'이고 '3등석'을 이용하며 '사망'
tt.query('''sex == "male" and age < 20 and pclass == 3 and survived == 0''')
```

```python
# '여성'이며 '2등석'을 이용하고 '생존'하며 '30세 이상 40세 미만'이고 '요금이 20파운드 이상'
tt.query('''sex == "female" and pclass == 2 and survived == 1 and age >= 30 and age < 40 and fare >= 20''')
```
