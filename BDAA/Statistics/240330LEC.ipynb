```python
#numpy 내에서도 기초통계치를 계산하는 함수 존재
#statistics 기초통계치 계산하는 함수
# scipy 등 다양하게 있다. (실제 판다스로도 계산 가능하다.)
import numpy as np
import statistics
```

```python
# 데이터의 대푯값이라고 할 수 있는 평균외에 다양한 지표들 문법
# 평균을 구하는 함수
# 중위값 구하는 함수
# 분산
# 표준편차
# 최대,최소
# 범위수 (4분위수)
# 상관계수
data = np.array([1,2,4,5,6,7,8,9,10,10])
```

```python
# 넘파이의 다양한 문법
np.mean(data) # np.기초통계치 

np.median(data) # 미디안

np.std(data) # 표준편차

np.var(data)

np.min(data)

np.max(data)

np.argmax(data) # 최대값의 인덱스 반환
 
np.argmin(data) # 최소값의 인덱스 반환

np.percentile(data, 25) # 제 1사분위수 25%

np.percentile(data, 75) # 제 3사분위수 75%
```

# 중앙값
- 데이터를 크기 순서대로 나열할 때 정확하게 중앙에 위치한 값
- 홀수면 중앙이 나온다, 짝수면 두 개의 값을 평균으로 계산한다.

- 중앙값은 평균값에 비해서는 이상치에 영향을 덜 받는다. 
- 평균으로 계산하는 게 아니라, 중앙값으로 대치해서 MSE 바꿔서 중앙값으로 바꿔서 계산하고 하는 이유가
- 데이터 자체에 이상치가 큰 경우 아무래도 대푯값을 평균보다는 중앙값이 적절하게 볼 수 있다. ( 도메인적인 부분과 같이 고려해야 한다.)

# 편차
편차는 각 데이터가 평균으로부터 얼마나 떨어져 있는지 지표
# 분산
산포도의 지표로 데이터 평균간의 차이를 이용, 분산은 편차의 제곱을 이용하여 계산
# 표준편차
산포도 지표로 분산에 제곱근을 취한 것을 표준편차

```python
#간단한 편차 계산
df_eng_mean=np.mean(df.english)
df['eng_dv']=df.english - df_eng_mean
```

```python
# 편차와 같은 값이 나온다
np.mean(df['eng_dv']**2)
```

```python
# 편차와 같은 값이 나온다
np.var(df.english)
```

# 사분위 범위, 범위

- 범위는 range
- 분산이나 표준편차와는 다르게 데이터를 전체를 보는 게 아니다.
- 최댓값, 최솟값으로만 산포도를 표현하는 것
- 최댓값과 최솟값이 차이가 크면 산포도가 크고, 차이가 작으면 작다.
- Rg = X.max - X.min

```python
np.max(df.english)- np.min(df.english)
```

- 사분위 범위
- 범위도 결국, 최대, 최소로 계산하는 것 -> 이상치가 크면 변화가 크다.
- 사분위 범위 interquartile range
- 데이터를 1로 보면 하위 25%, 중위 50%, 상위 75% 
- 제1사분위수, 제2사분위수(50%), 제3사분위수 Q1, Q2, Q3
- 사분위의 범위 IQR
- IQR = Q3- Q1

# 데이터의 정규화
- normalization 데이터의 정규화
- 동일한 시험 점수더라도 평균이 30점인 시험에서 (어려운 시험) 60점 받은 것
- 평균이 90점 인데 60점 받은 것 (상대적으로 쉬운 시험) 
- 평균이나 분산에 따라서 값이 달라진다.

- 평균이나, 분산에 의존하지 않고 데이터의 상대적인 위치관계를 알 수 있는 지표는 무엇인가?
- 가장 대표적인 게 편찻값 
- 평균이나 분산이 어떤 값이던 편차값도 결국 더 통일된 지표가 필요하게 되어서
- 정규화를 통해서 다양한 데이터의 기준들을 맞추자
