```python
# 데이터 병합

# 두 개의 데이터 프레임을 붙인다
# 두 개를 붙여서 우리가 확인해야 하는 상황
# 고객, 주문
# 하나의 테이블은 고객에 대한 정보, 나머지 하나는 주문에 대한 정보
# 두 개의 테이블을 붙여서 고객과 주문에 대해서 같이 볼 수 있는 것
```

```python
# 판다스에서 제공하는 문법

# merge()
# 기준 컬럼을 잡고 병합을 하는 경우, 공동의 컬럼을 하나 잡고 병합하고
# left, right outer, inner 등 조인 방법이 다양하다
# 공통된 컬럼은 컬럼 이름이 아니라 컬럼 내에 값을 의미하는 것
# 고객 아이디라는 컬럼이 두 개 동일한데, 하나는 abcd1234, 다른 하나는 -123123123 숫자로 암호화가 되어있느면 두 개는 공통된 컬럼이라 볼 수 없다

# concat()
# df1, df2 두 개의 데이터 프레임이 있다면 행 기준, 열 기준으로 붙인다
# df1 + df2 두 개로 덩어리 형태로 붙인다. (공통된 기준 컬럼이 없다.)
```

```python
test1=pd.DataFrame({'학회원번호':[1,2,3,4,5],'학회원분반':['Python','SQL','R','데이터분석입문','데이터분석전처리']})
test1
```

```python
test2=pd.DataFrame({'학회원번호':[1,2,3,4,5],'상점현황':[4,3,5,10,2]})
test2
```

```python
# pd.merge(왼쪽시작컬럼(기준),오른쪽 컬럼(기준 옆 붙일),how='병합방법',on='공통컬럼지정'(left_on, right_on))
display(test1)
display(test2)
pd.merge(test1,test2,how='inner')
```

```python
# pd.concat([df1, df2, df3....])
pd.concat([test1,test2]) #디폴트는 아래로 붙임  0기준이 디폴트 
```

```python
pd.concat([test1,test2],axis=1) # 디폴트는 아래로 붙임
# Axis 축을 지정하여 행, 열로 잡음
# 0은 행
# 1은 열
```

```python
pd.concat([test1,test2,test1],axis=1)
```

```python
# 타이타닉데이터를 가지고 시각화를 통해서 여러가지 컬럼들을 비교해 보자!

# EDA, 탐색적으로 데이터를 바라본다. 거기서 인사이트를 찾는다.
# 머신러닝 진행할 예정 -> 머신러닝은 결국 패턴 가지고 학습하는 것
# 생존과 생존하지 못한 데이터를 가지고 둘의 차이를 학습하고, 그 차이는 여러 컬럼에 따라 달라지는 것
# 그 컬럼들을 배우면서 -> 이러한 컬럼들의 패턴을 보이면 생존이다., 아니면 생존하지 못했다. 예측
# ML쪽은 y값(생존)에 따른 다른 피처들의 패턴이 정말 중요하다.
# 시각적으로 바라보거나, 기초통계적인 것으로도 봤을 때 y값과 다른 피처들의 관계가 정말 중요하다.
# ML 지도학습 -> 정답가지고 패턴 학습, 정답에 대한 패턴을 머신에게 얼마나 잘 알려주는지에 따라 성능도 올라간다.
# 기본적인 비즈니스 로직, 어떤 y값과 피처들의 로직이 명확하면 해당 피처를 사용해야 한다.
# 영업이익 계산하면-> 매출이런 것들 중요한 피처다.
# 머신에게 학습시킬 것들 중 정답에 대한 비중이 차이가 많이 나는 경우는 해당 피처를 사용할 수도 있다.
# A라는 피처와 B라는 피처가 있다.
# A는 정답 유뮤를 7:3으로 나눠준다. B는 5:5로 나눠준다. (Y값을 나눌 때)
# y값에 대해서 피처들이 어떤 식으로 나눠주는지, 패턴, 분포들이 보여지는지 이런 부분을 꼭 확인하고 작업을 해야 한다.
```

```python
# Y값들과의 관계에 따른 다른 피처들 확인하자!
df[['pclass','survived']].groupby(['pclass']).mean().sort_values(by='survived',ascending=False)
```

```python
df[['sex','survived']].groupby(['sex']).mean().sort_values(by='survived',ascending=False)
```
