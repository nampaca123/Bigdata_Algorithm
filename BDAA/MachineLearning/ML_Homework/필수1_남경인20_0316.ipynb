```python
import seaborn as sns
tt = sns.load_dataset('titanic')
```

```python
# '생존'하며 '여성'이고 '25세 미만'
tt.query('''sex == "female" and survived == 1 and age < 25 ''')
```

```python
# '사망'이며 '남성'이고 '40세 이상'
tt.query('''sex == "male" and survived == 10 and age > 40 ''')
```

```python
# '3등석'이며 '생존'이고 '18세 이상 60세 이하'
tt.query('''pclass == 3 and survived == 1 and age >= 18 and age <= 60''')
```

```python
# '1등석'이며 '사망'이고 '여성'
tt.query('''pclass == 1 and survived == 0 and sex == "female"''')
```

```python
# '남성'이며 '퀸즈타운에서 출발'이고 '2등석'
tt.query('''sex == "male" and embarked == "Q" and pclass == 2''')
```

```python
# '생존'이며 '쉐르부르에서 출발'이고 '요금이 100 이상'
tt.query('''survived == 1 and embarked == "C" and fare > 100''')
```

```python
# '여성'이며 '사망'이고 '사우스햄튼에서 출발'
tt.query('''sex == "female" and survived == 0 and embarked == "S"''')
```

```python
# '40세 이상'이며 '남성'이고 '3등석'
tt.query('''age > 40 and sex == "male" and pclass == 3''')
```

```python
# '18세 미만'이며 '생존'이고 '쉐르부르에서 출발'
tt.query('''age < 18 and survived == 1 and embarked == "C"''')
```

```python
# '사망'이며 '50세 이상'이고 '1등석'
tt.query('''survived == 0 and age > 50 and pclass == 1''')
```
