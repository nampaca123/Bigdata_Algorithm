```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tt = sns.load_dataset('titanic')
```

```python
tt['age_group'] = pd.cut(tt['age'], bins=[0, 10, 20, 40, 100], labels=['10세 미만', '10대', '20대와 30대', '그 외'])
# 나이 그룹에 따라 구분하기 위한 파생변수 형성
# bins : pd.cut을 이용해서 데이터를 나누고자 하는 구간의 경계값 정의
# labels: 경계값을 기준으로 하여 정의된 구간에 이름을 부여함
```

```python
grouped_fares = tt.groupby(['sex', 'age_group'])['fare'].mean().reset_index()
#나이 그룹과 성별을 기준으로 한 평균 요금 계산
# reset_index() : 인덱스를 다시 데이터프레임으로 전환하여 도표화 가능하게 만듬
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='fare', hue='sex', data=grouped_fares)
plt.title('나이 그룹 및 성별 기준 평균 요금')
plt.xlabel('나이 그룹')
plt.ylabel('평균 요금')
plt.legend(title='성별')
plt.grid(True)
plt.show()
```
