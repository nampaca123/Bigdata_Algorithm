```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset('titanic')
```

```python
# 1. 가설 1) 성별에 따른 생존율의 차이
# 여성이 남성보다 생존율이 높을 것이다.

gender_survival = titanic.groupby('sex')['survived'].mean()
plt.figure(figsize=(10,6))
sns.barplot(x=gender_survival.index, y=gender_survival.values, palette='pink')
plt.title('성별에 따른 생존율의 차이')
plt.xlabel('성별')
plt.ylabel('생존율')
plt.show()
```

```python
# 2. 가설 2) 탑승한 도시별 객실 등급에 따른 생존율 차이
geksil = sns.FacetGrid(titanic, row='embark_town')
geksil.map(sns.pointplot, 'class', 'survived')
geksil.set_xlabels('객실 등급')
geksil.set_ylabels('생존율')
```

```python
# 3. 가설 ) 가족 동반 여부에 따른 생존율 차이
# 가족 동반 여부 생성
titanic['with_family'] = (titanic['sibsp'] + titanic['parch']) > 0
# 타이타닉 데이터셋에서 sibsp는 타이타닉에 탑승한 형제자매, 배우자의 수를 나타내며
# parch는 같은 배에 탑승한 부모나 자녀의 수를 나타낸다. 따라서 이 두 값을 기준으로 가족과의 동승유무를 판단할 수 있다.
family_survival = titanic.groupby('with_family')['survived'].mean()
fam = sns.catplot(x="with_family", y="survived", data=titanic, kind="bar")
fam.set_titles('가족 동반 여부에 따른 생존율의 차이')
fam.set_xlabels('가족 동반 여부')
fam.set_ylabels('생존율')
```
