```python
# 새로운 파생변수를 만들어보자
# 파생변수 = 기존 변수를 이용해서 새로운 변수를 만드는 것 
# 기존 변수를 합치거나 둘 이상을, 기존 변수에 새로운 값을 사칙연산을 통해 계산하거나 등등 

# 요금에 단위를 맞추기 위해 100을 다 곱하자!
# 새로운 변수에 값을 넣어야 한다.
# 변수를 만들어야 한다.
# 변수를 만들기 위해서는 데이터프레임에서 변수를 만들어야 한다.

# 파생변수 1차로 만들기
tt['fare_100'] = 0 #초깃값 세팅 python 문법
tt['fare_100'] = tt.fare * 100 #변수를 만들어서 새로운 값을 넣을 수 있다.
```

```python
# if문과 for문을 이용한 전처리 진행
# Q1: 남성과 여성의 fare 평균을 새로운 변수로 만들고 싶다!
# A1: 내가 원하는 칼럼은 fare에 대한 남성과 여성에 따른 평균값이 다른 fare 컬럼의 값이 매핑된다

# 남성의 평균 요금
me_fa_mean = tt.query('''sex == "male"''')['fare'].mean()

# 여성의 평균 요금
fe_fa_mean = tt.query('''sex == "female"''')['fare'].mean()
```

```python
tt['female_fare_mean'] = 0
tt['male_fare_mean'] = 0

for i in range(len(tt)):
    #if문을 이용해서 남성과 여성의 값을 비교하고, 파생변수를 만들어서 그곳에 남성과 여성의 평균 fare값을 넣는 변수를 만들자.
    if tt.loc[i]['sex'] == 'female':
        tt['female_fare_mean'].loc[i] = fe_fa_mean
        
    elif tt.loc[i]['sex'] == 'male':
        tt['male_fare_mean'].loc[i] = me_fa_mean
```

```python
# assign(원하는 요구 조건, 넘파이를 이용해서 where를 사용할 예정 )
import numpy as np

tt.assign(age_old = np.where(tt['age']>30, 'old','young'))
# assign(파생변수명 = np.where(요구조건, 파생변수값, 파생변수 나머지값))
```

```python
# np.where 해당 조건의 인덱스를 반환한다.
np.where(tt['age']>30)
```

```python
tt_new=tt.assign(age_old = np.where(tt['age']>30, 'old','young'))
tt_new
```

```python
tt.assign(fare_100 = lambda x: x['fare']*100) # assign lambda 함수도 쉽게 넣어서 파생변수를 만들 수 있다.
```
