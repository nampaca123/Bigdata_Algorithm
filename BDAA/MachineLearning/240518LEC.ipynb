# age 컬럼 전처리
- Missing Value 대체하는 방법
- MICE
- 보간법
- 단순한 기초통계치로 대체하는 법
- KNN, ML 기법을 통해서 대체하는 법 

```python
grid = sns.FacetGrid(train_df, row='Pclass', col='Sex', aspect =1.5)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
```

- 단순하게 Age 평균으로 대체하면 -> Median 
- Pclass, Sex 따라서 2차원으로 나눈 후 -> 그 값들의 Median, Mean 대체를 한다.

```python
guess_ages = np.zeros((2,3)) # np.zeros 0값 넣는다. 일단 na값을 채워넣을 주머니 하나 만들었다.
```

```python
## 결측치를 채울 코드는?
## pclass, 성별 2 두가지의 차원으로 결측치를 바라볼 것
## 두 가지의 피처로 값을 추출하면서 이 값의 median의 값을 -> guess_ages 넣어서 대체하면 된다.
## age는 22.1살 22.7살 없다. 반올림을 하는 작업을 진행해야 한다.

for df in combine:
    for i in range(0,2):
        for j in range(0,3):
            guess_df = df[(df['Sex']==i) & (df['Pclass']==j+1)]['Age'].dropna()
            
            age_guess = guess_df.median()
            guess_ages[i,j] = (age_guess/0.5 + 0.5) * 0.5
            
            
    for i in range(0,2):
        for j in range(0,3):
            df.loc[(df.Age.isnull()) & (df.Sex ==i)&(df.Pclass==j+1), 'Age'] = guess_ages[i,j]
    df['Age'] = df['Age'].astype(int)
```

## 결측치의 대치 방법들

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np

# 캘리포니아 주택 데이터셋 로드
california = fetch_california_housing()
data = pd.DataFrame(california.data, columns=california.feature_names)
data['MedHouseVal'] = california.target

# 'MedInc' 변수를 독립 변수로 사용
data = data[['MedInc', 'MedHouseVal']]
```

```python
data
```

- 결측치를 생성하여서
- 실제값과, 결측치를 다양하게 대치한 값들의 비교를 통해 인사이트를 확인하자!

```python
#결측치를 생성

#결측치를 10%
np.random.seed(42)
missing_rate = 0.1
n_ms_sample = int(np.floor(missing_rate*data.shape[0]))


missing_samples = np.random.choice(data.index, n_ms_sample, replace=False)
data_missing = data.copy()
data_missing.loc[missing_samples, 'MedInc' ] = np.nan
```
