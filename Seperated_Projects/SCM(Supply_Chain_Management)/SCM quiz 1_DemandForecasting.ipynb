```python
import pandas as pd
```

```python

# 'n' value for Moving Average is 3
# Alpha value for Exponential Smoothing is 0.35
# Weighted Moving Average is W1=0.5, W2=0.4, W3=0.1
# Alpha value for Trend Adjusted Exponential Smoothing is 0.4
# Beta value for Trend Adjusted Exponential Smoothing is 0.5
Demand = [125,220,310,450,630,450,400,350,380,410]
Value = []
```

```python
# Naive Model

print("※ Naive Model\n")
for i in range(10):
    if i-1<0: 
        Value.append(0)
        print("Time:",i+1,"Value:", Value[i])
    else:
        Value.append(Demand[i-1])
        print("Time:",i+1,"Value:",Value[i])
Value.clear()
```

```python
# Moving Average

print("※ Moving Average\n")
for i in range(10):
    if i<3: 
        Value.append(0)
        print("Time:",i+1,"Value:", Value[i])
    else:
        Value.append((Demand[i-1]+Demand[i-2]+Demand[i-3])/3)
        print("Time:",i+1,"Value:", round(Value[i],2))
        
print("\n※ MAD\n")
madsum = 0
for i in range(3,10):
    madsum += (abs(Value[i]-Demand[i]))
mad = madsum/7
print("MAD =",round(mad,2))
Value.clear()
```

```python
# Weighted Moving Average

print("※ Weighted Moving Average\n")
for i in range(10):
    if i<3: 
        Value.append(0)
        print("Time:",i+1,"Value:", Value[i])
    else:
        Value.append((0.5*Demand[i-1]+0.4*Demand[i-2]+0.1*Demand[i-3]))
        print("Time:",i+1,"Value:", round(Value[i],2))
        
print("\n※ MAD\n")
madsum = 0
for i in range(3,10):
    madsum += (abs(Value[i]-Demand[i]))
mad = madsum/7
print("MAD =",round(mad,2))
Value.clear()
```

```python
# Exponential Smoothing

print("※ Exponential Smoothing\n")

for i in range(10):
    if i == 0 :
        Value.append(100)
        print("Time:",i+1,"Value:", Value[i])
    else:
        Value.append((0.35*Demand[i-1]+0.65*Value[i-1]))
        print("Time:",i+1,"Value:", round(Value[i],2))
        
print("\n※ MAD\n")
madsum = 0
for i in range(10):
    madsum += (abs(Value[i]-Demand[i]))
mad = madsum/10
print("MAD =",round(mad,2))
Value.clear()
```

```python
# Estimate of Level
# Alpha value for Trend Adjusted Exponential Smoothing is 0.4
# Beta value for Trend Adjusted Exponential Smoothing is 0.5

print("※ Estimate of Level\n")

LValue=[]
TValue=[]
Timelist=[]
for i in range(10):
    if i == 0 :
        LValue.append(30)
        TValue.append(15)
        Timelist.append(i+1)
        print("Time:",i+1,"A(Estimate of level) Value:", round(LValue[i],2),"T(Estimate of trend) Value:",round(TValue[i],2))
    elif i<9:
        LValue.append(0.4*Demand[i]+(1-0.4)*(LValue[i-1]+TValue[i-1]))
        TValue.append(0.5*(LValue[i]-LValue[i-1])+(1-0.5)*(TValue[i-1]))
        Timelist.append(i+1)
        print("Time:",i+1,"A(Estimate of level) Value:", round(LValue[i],2),"T(Estimate of trend) Value:",round(TValue[i],2))
    else:
        LValue.append(0)
        TValue.append(0)
        Timelist.append(i+1)
        print("Time:",i+1,"A(Estimate of level) Value: -","T(Estimate of trend) Value: -")
   
eol=pd.DataFrame({
    'Time':Timelist,
    'A(Estimate of level) Value':LValue,
    'T(Estimate of trend) Value':TValue
    })
eol = eol.round(2)
eol['A(Estimate of level) Value'] = eol['A(Estimate of level) Value'].astype(object)
eol['T(Estimate of trend) Value'] = eol['T(Estimate of trend) Value'].astype(object)
eol.at[9, 'A(Estimate of level) Value'] = "-"
eol.at[9, 'T(Estimate of trend) Value'] = "-"
eol
```

```python
# Trend Adjusted Exponential

print("※ Trend Adjusted Exponential\n")

for i in range(10):
    if i == 0 :
        Value.append(0)
        print("Time:",i+1,"Value: -")
    else:
        Value.append(LValue[i-1]+TValue[i-1])
        print("Time:",i+1,"Value:", round(Value[i],2))
        
print("\n※ MAD\n")
madsum = 0
for i in range(10):
    madsum += (abs(Value[i]-Demand[i]))
mad = madsum/9
print("MAD =",round(mad,2))
Value.clear()
```
